```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
%matplotlib inline
```

```python
df = pd.read_csv('/Users/jon/Documents/IRONHACK/PROJECTS/PROJECT4/beer_reviews_clean.csv')
df.head()
```

```python
df['year'] = pd.DatetimeIndex(df['review_time']).year
df.head()
```

```python
df['month'] = pd.DatetimeIndex(df['review_time']).month
df.head()
```

```python
df['date'] = pd.to_datetime(df['review_time']).dt.date
df.head()
```

```python
df['day_of_week'] = df['date'].apply(lambda x: datetime.datetime.strftime(x, '%A'))
df.head()
```

```python
df['beer_style'].value_counts()
```

```python
plt.hist(df['review_overall'], bins = 10)
plt.title("Distribution of Scores for all Beers")
plt.xlabel("Score")
plt.ylabel("Score Count")
```

```python
plt.figure(figsize=(10,10))
plt.title("Score Distribution for Contributing Factors")
plt.subplot(2,2,1)
plt.hist(df['review_aroma'])
plt.subplot(2,2,2)
plt.hist(df['review_appearance'])
plt.subplot(2,2,3)
plt.hist(df['review_palate'])
plt.subplot(2,2,4)
plt.hist(df['review_taste'])
#Struggling to plot labels and title for each graph
```

```python
plot_order = df.groupby('beer_style')['review_overall'].mean().sort_values(ascending=False).index.values
plot = sns.catplot(data=df, x='beer_style',  y='review_overall',kind='bar',ci=None, legend_out=False, order=plot_order[1:104], aspect=3, height=3)
plot.set_xticklabels(rotation=90, horizontalalignment='right').ax.tick_params(labelsize=5)
plt.grid()
plt.title('Mean Score by Beer Style', size='15', fontweight='bold')
plt.xlabel("Beer Style")
plt.ylabel("Mean Score")
```

```python
plot_order = df.groupby('beer_style')['review_overall'].mean().sort_values(ascending=False).index.values
plot = sns.catplot(data=df, x='beer_style',  y='review_overall',kind='bar',ci=None, legend_out=False, order=plot_order[1:6], aspect=3, height=3)
plot.set_xticklabels(rotation=90, horizontalalignment='right').ax.tick_params(labelsize=15)
plt.grid()
plt.title('Best Mean Score by Beer Style', size='15', fontweight='bold')
plt.xlabel("Beer Style")
plt.ylabel("Mean Score")
```

```python
plot_order = df.groupby('beer_style')['review_overall'].mean().sort_values(ascending=False).index.values
plot = sns.catplot(data=df, x='beer_style',  y='review_overall',kind='bar',ci=None, legend_out=False, order=plot_order[-6:], aspect=3, height=3)
plot.set_xticklabels(rotation=90, horizontalalignment='right').ax.tick_params(labelsize=15)
plt.grid()
plt.title('Lowest Mean Score by Beer Style', size='15', fontweight='bold')
plt.xlabel("Beer Style")
plt.ylabel("Mean Score")
```

```python
values = ["northyorksammy","BuckeyeNation", "mikesgroove", "Thorpe429", "womencantsail", "NeroFiddled", "ChainGangGuy", "brentk56", "Phyl21ca", "WesWes"]
years = [2005,2006,2007,2008,2009,2010,2011]
df_top_users =  df.loc[(df["review_profilename"].isin(values))]
df_top_users = df_top_users.groupby('review_profilename')['year'].value_counts()
df_top_users = pd.DataFrame(df_top_users)
df_top_users.rename(columns = {"year":"count"}, inplace=True)
df_top_users = df_top_users.rename_axis(['Name', 'Year']).reset_index()
df_top_users = df_top_users.loc[(df_top_users["Year"].isin(years))]
df_top_users
```

```python
plt.figure(figsize=(14, 5))

sns.lineplot(data = df_top_users,
             x = df_top_users["Year"],
             y = df_top_users['count'],
             hue = df_top_users["Name"],
             ci=None
             )

plt.title("Number of Reviews per Year for the Top 10 Users")
plt.legend(loc='best')
plt.xlabel("Year")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
plt.figure(figsize=(14, 5))

sns.lineplot(data = df_top_users,
             x = df_top_users["Year"],
             y = df_top_users['count']/365,
             hue = df_top_users["Name"],
             ci=None
             )

plt.legend(loc='best')
plt.xlabel("Year")
plt.ylabel("Number of Reviews/Day")
plt.title("Reviews/Day/year for the Top 10 Users")
plt.show()
```
